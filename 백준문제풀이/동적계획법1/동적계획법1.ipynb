```python
import sys
#input = sys.stdin.readline
dp = [[[1 for _ in range(51)] for _ in range(51)] for _ in range(51)]
for c in range(1, 51):
    for b in range(1 ,51):
        for a in range(1, 51):
            if a < b and b < c:
                dp[c][b][a] =  dp[c-1][b][a]+ dp[c-1][b-1][a] - dp[c][b-1][a]
            else:
                dp[c][b][a] = dp[c][b][a-1]+ dp[c][b-1][a-1] + dp[c-1][b][a-1] - dp[c-1][b-1][a-1]
                
while True:
    a, b, c = map(int, input().split())
    if a==-1 and b==-1 and c==-1:
        break
    if a <= 0 or b<=0 or c<= 0:
        answer = 1
    elif a>= 20 or b>= 20 or c >= 20:
        answer = dp[20][20][20]
    else:
        answer= dp[c][b][a]
    print(f'w({a}, {b}, {c}) = {answer}')
    
```

```python
import copy
n = int(input())
dp = [0]*n
nums = []
for _ in range(n):
    nums.append(list(map(int, input().split())))

dp = [0]*n
new_dp = [0]*n
for num in (nums):
    for idx2, value in enumerate(num):
        if idx2 == 0:
            new_dp[idx2] = dp[idx2]+value
        elif idx2 == len(num)-1:
            new_dp[idx2] = dp[idx2-1]+value
        else:
            new_dp[idx2]= max(dp[idx2-1]+value, dp[idx2]+value)
    dp = copy.deepcopy(new_dp)
print(max(dp))
```
