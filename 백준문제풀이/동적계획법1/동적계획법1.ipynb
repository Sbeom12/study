```python
n = int(input())
fibo = [0] * 41
fibo[1], fibo[2] = 1,1 
for i in range(3, 41):
    fibo[i] = fibo[i-1]+ fibo[i-2]

print(fibo[n], n-2) # 더하기 연산은 항상 n-2임.
```

# 신나는 함수 실행.
---


* 문제 조건
    * 아무거나 0보다 작거나 같으면 1
    * 아무거나 20보다 크면 W(20, 20, 20)
    * a < b < c 순서라면
        * $w(a, b, c)= w(a,b,c-1) +w(a, b-1, c-1) - w(a, b-1, c)$
    * 나머지는
        * $w(a, b, c)=  w(a-1, b, c) + w(a-1, b-1, c) + w(a-1, b, c-1) - w(a-1, b-1, c-1) $

```python
dp = [[[1 for _ in range(51)] for _ in range(51)] for _ in range(51)]
```

```python
import sys
#input = sys.stdin.readline
dp = [[[1 for _ in range(51)] for _ in range(51)] for _ in range(51)]
for c in range(1, 51):
    for b in range(1 ,51):
        for a in range(1, 51):
            if a < b and b < c:
                dp[c][b][a] =  dp[c-1][b][a]+ dp[c-1][b-1][a] - dp[c][b-1][a]
            else:
                dp[c][b][a] = dp[c][b][a-1]+ dp[c][b-1][a-1] + dp[c-1][b][a-1] - dp[c-1][b-1][a-1]
                
while True:
    a, b, c = map(int, input().split())
    if a==-1 and b==-1 and c==-1:
        break
    if a <= 0 or b<=0 or c<= 0:
        answer = 1
    elif a>= 20 or b>= 20 or c >= 20:
        answer = dp[20][20][20]
    else:
        answer= dp[c][b][a]
    print(f'w({a}, {b}, {c}) = {answer}')
    
```
