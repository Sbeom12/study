* 문제 조건
    * 아무거나 0보다 작거나 같으면 1
    * 아무거나 20보다 크면 W(20, 20, 20)
    * a < b < c 순서라면
        * $w(a, b, c)= w(a,b,c-1) +w(a, b-1, c-1) - w(a, b-1, c)$
    * 나머지는
        * $w(a, b, c)=  w(a-1, b, c) + w(a-1, b-1, c) + w(a-1, b, c-1) - w(a-1, b-1, c-1) $

```python
dp = [[[1 for _ in range(51)] for _ in range(51)] for _ in range(51)]
```

```python
import sys
#input = sys.stdin.readline
dp = [[[1 for _ in range(51)] for _ in range(51)] for _ in range(51)]
for c in range(1, 51):
    for b in range(1 ,51):
        for a in range(1, 51):
            if a < b and b < c:
                dp[c][b][a] =  dp[c-1][b][a]+ dp[c-1][b-1][a] - dp[c][b-1][a]
            else:
                dp[c][b][a] = dp[c][b][a-1]+ dp[c][b-1][a-1] + dp[c-1][b][a-1] - dp[c-1][b-1][a-1]
                
while True:
    a, b, c = map(int, input().split())
    if a==-1 and b==-1 and c==-1:
        break
    if a <= 0 or b<=0 or c<= 0:
        answer = 1
    elif a>= 20 or b>= 20 or c >= 20:
        answer = dp[20][20][20]
    else:
        answer= dp[c][b][a]
    print(f'w({a}, {b}, {c}) = {answer}')
    
```

```python
N = int(input())
dp = [0] * (N+1)
dp[1] = 1
dp[2] = 2
for i in range(3, N+1):
    dp[i] = dp[i-1]+dp[i-2]
print(dp[N]%15746)
```

* 함정
---
* python의 int는 4바이트가 아닙니다. 기본적으로만 몇십 바이트를 사용하며, 표현하는 수의 크기가 커질수록 자릿수에 비례하는 만큼 메모리를 더 차지합니다.

* 이 코드의 점화식은 n번째 항의 자릿수가 O(n)이기 때문에 n개의 수를 모두 개별적으로 저장하면 필요한 메모리는 O(n^2)이 됩니다.

```python
# 메모리 할당량 확인
import sys
from matplotlib import lines
import matplotlib.pyplot as plt

x = []
y = []
for i in range(100):
    x.append(i)
    y.append(sys.getsizeof(2**i))

plt.plot(x, y, color='black')
plt.xticks(range(0, 100, 30))
plt.xlabel('X')
plt.ylabel('sizeof(2^X)(byte)')
for i in range(1, 100, 30):
    plt.vlines(i, min(y), max(y), color='red', linestyle='--')
plt.show()
```

```python
import copy
n = int(input())
dp = [0]*n
nums = []
for _ in range(n):
    nums.append(list(map(int, input().split())))

dp = [0]*n
new_dp = [0]*n
for num in (nums):
    for idx2, value in enumerate(num):
        if idx2 == 0:
            new_dp[idx2] = dp[idx2]+value
        elif idx2 == len(num)-1:
            new_dp[idx2] = dp[idx2-1]+value
        else:
            new_dp[idx2]= max(dp[idx2-1]+value, dp[idx2]+value)
    dp = copy.deepcopy(new_dp)
print(max(dp))
```
