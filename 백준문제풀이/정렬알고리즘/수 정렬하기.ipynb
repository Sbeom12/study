```python
N = int(input())
n_list = [int(input()) for _ in range(N)]
n_list.sort()
for num in n_list:
    print(num)
```

# 수 정렬하기 3
---



## 검토
---
int형 = 4byte
* 제한 : 8MB = 8,000,000 byte
    * 최대 2,000,000 개
    * sort 사용시 N개의 메모리 공간 필요.
        * 메모리 초과 발생.

```python
import sys
num_list = [0] * 10001
num = int(sys.stdin.readline())

for i in range(num):
    input_num = int(sys.stdin.readline())
    num_list[input_num] = num_list[input_num] + 1 # 각 인덱스가 몇번 나왔는지 확인.

# 3. 정렬된 정수를 한 줄에 하나씩 출력한다.
for i in range(10001):
    if num_list[i] == 0:
        continue
    for j in range(num_list[i]):
        print(i)
```

# 소트인사이드

```python
num = list(map(int, list(input())))
num.sort(reverse=True) # 내림차순으로 정렬
for i in num:
    print(i, end='') # 한줄로 출력을 위해서 end를 \n에서 ''로 변환
```

# 좌표정렬하기
---


```python
import sys
input = sys.stdin.readline
N= int(input())
dot_list = [list(map(int, input().split())) for _ in range(N)]
dot_list.sort(key=lambda x:(x[0], x[1]))
for dot in dot_list:
    print(*dot)
```

# 좌표 정렬하기 2

```python
import sys
num = int(sys.stdin.readline())
num_list = [list(map(int, sys.stdin.readline().split())) for _ in range(num)]
num_list.sort(key=lambda x:(x[1], x[0]))
for num in num_list:
    print(*num)
```

# 단어 정렬
---


```python
import sys
#input = sys.stdin.readline
N = int(input())
word_list = []
for _ in range(N):
    word = input()
    if [word, len(word)] in word_list:
        continue
    word_list.append([word, len(word)])
word_list.sort(key=lambda x:(x[1], x[0]))

for word in word_list:
    print(word[0])
```

# 나이순 정렬
---


```python
import sys
num = int(sys.stdin.readline())
a = [tuple(sys.stdin.readline().split()) for _ in range(num)]
a.sort(key=lambda x:int(x[0])) # 첫 번째 요소인 나이를 기준으로 정렬 및 int형 으로 변환해서 sort
for value in a:
    print(value[0], value[1])
```

# 좌표 압축
---


```python
import sys
input = sys.stdin.readline
N= int(input())
num_list = list(map(int, input().split()))
temp_dict ={num:idx for idx, num in enumerate(sorted(set(num_list)))}

for num in num_list:
    print(temp_dict[num], end=' ')
```

## 검토
---
* 리스트의 경우 탐색하려는 범위가 크면 클수록 탐색시간이 증가하게됨.
* 딕셔너리의 경우에 hash table을 이용.
    * Key값에 따라 value가 저장될 위치가 결정되어, 배열의 전체를 탐색하지 않고도 value를 얻을 수 있음.
* 10,000,000를 찾을 때 딕셔너리가 보통 20배 빠름.
