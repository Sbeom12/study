```python
from collections import deque
N, M = map(int, input().split())
graph = []
for _ in range(N):
    graph.append(list(map(int, input())))

#BFS
queue = deque()
queue.append((0,0))
total_num = [[0 for _ in range(M)] for _ in range(N)]
total_num[0][0] = 1
move_sequence = [(-1,0), (1, 0), (0,1), (0,-1)] #   .

while queue:
    x, y = queue.popleft()
    condition = x < 0 or y < 0 or x >=M or y >= N
    #   
    if condition: 
        continue
    if graph[y][x] == 1:
        graph[y][x] = 0
        #     
        for i in range(4):
            nx, ny = x + move_sequence[i][0], y+ move_sequence[i][1]
            if nx < 0 or ny < 0 or nx>= M or ny >= N:
                continue
            if graph[ny][nx] == 1: #    
                total_num[ny][nx] = total_num[y][x] +1
                queue.append((nx, ny))
print(total_num[-1][-1])
```
