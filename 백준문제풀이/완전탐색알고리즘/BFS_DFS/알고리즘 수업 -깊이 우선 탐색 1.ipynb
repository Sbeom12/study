# 알고리즘 수업 - 깊이 우선 탐색 1
---


## 풀이
* 시간 제한 1초 : 계산량 대충 2000만 정도
* 메모리 제한 512MB
* DFS
    * 구현 방법 : stack or 재귀함수

```python
from collections import defaultdict
import sys
N, E, R = map(int, sys.stdin.readline().split())
graph = defaultdict(set)
for _ in range(E):
    node1, node2 = map(int, sys.stdin.readline().split())
    graph[node1].add(node2)
    graph[node2].add(node1) 

sequence = [0 for _ in range(N+1)]
visited = [False for _ in range(N+1)] # 방문 확인.
count = 1
stack = [R]
while stack:
    current=stack.pop() # 현재 방문
    if visited[current]: # 방문 했다면 pass
        continue
    sequence[current] = count
    visited[current] = True
    for n in reversed(sorted(graph[current])):
        if visited[n]: # 이전에 방문햇으면 통과
            continue
        stack.append(n)
    count +=1
for answer in sequence[1:]:
    print(answer)
```

## 검산: 
---
* N = 100,000
* M = 200,000
* 정렬사용시 : $O(Nlog(N))$  
* 인접행렬시 : $O(n^2)$  
* 인접리스트 : $O(N+M)$  

* 인접리스트 시 통과 가능.


## 특이점.
* 백준의 최대 재귀 깊이는 약 1000정도.
* 만약 문제가 1000보다 넘어가는 경우에는 조치가 필요
    * `sys.setrecursionlimit`를 조절해야함.
* 또한 연결된 노드에서 낮은 노드로 가야하므로 이를 해결하기 위해 sort를 사용해야함.
