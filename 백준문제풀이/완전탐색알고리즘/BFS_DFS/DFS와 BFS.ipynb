# DFS와 BFS
---


## 풀이
* 시간 제한 2초 : 계산량 대충 4000만 정도
* 메모리 제한 128MB
* DFS
    * 구현 방법 : stack or 재귀함수
* BFS
    * 구현 방법 : queue

```python
from collections import defaultdict, deque
N, E, V =  map(int, input().split())
graph = defaultdict(set)
for _ in range(E):
    node1, node2 = map(int, input().split())
    graph[node1].add(node2)
    graph[node2].add(node1) 
visited = [0 for _ in range(N+1)] # 방문 확인.
sequence_bfs = []
sequence_dfs = []

# BFS
queue = deque() 
queue.append(V) 
while queue:
    current=queue.popleft() # 현재 방문
    if visited[current] == 1: # 방문 했다면 pass
        continue
    visited[current] = 1
    sequence_bfs.append(current)
    for n in sorted(graph[current]):
        if visited[n] == 1: # 이전에 방문햇으면 통과
            continue
        queue.append(n)
# DFS
stack = [V] # DFS 
visited = [0 for _ in range(N+1)] # 방문 확인.
while stack:
    current=stack.pop() # 현재 방문
    if visited[current]==1: # 방문 했다면 pass
        continue
    sequence_dfs.append(current)
    visited[current] = 1
    for n in reversed(sorted(graph[current])):
        if visited[n]==1: # 이전에 방문햇으면 통과
            continue
        stack.append(n)
print(*sequence_dfs)
print(*sequence_bfs)
```

## 검산: 
---
* 2초 
* N = 1,000
* M = 10,000
* 정렬사용시 : $O(Nlog(N))$  
* 인접행렬시 : $O(n^2)$  
* 인접리스트 : $O(N+M)$  
* $O(M+N+M+N+M)$ = $O(3M+2N)$

</br>

* 인접행렬 & 인접 리스트 둘다 가능.
    * 위 코드는 인접리스트와 유사
