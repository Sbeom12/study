# DFS BFS
---


## 
*   2 :   4000 
*   128MB
* DFS
    *   : stack or 
* BFS
    *   : queue

```python
from collections import defaultdict, deque
N, E, V =  map(int, input().split())
graph = defaultdict(set)
for _ in range(E):
    node1, node2 = map(int, input().split())
    graph[node1].add(node2)
    graph[node2].add(node1) 
visited = [0 for _ in range(N+1)] #  .
sequence_bfs = []
sequence_dfs = []

# BFS
queue = deque() 
queue.append(V) 
while queue:
    current=queue.popleft() #  
    if visited[current] == 1: #   pass
        continue
    visited[current] = 1
    sequence_bfs.append(current)
    for n in sorted(graph[current]):
        if visited[n] == 1: #   
            continue
        queue.append(n)
# DFS
stack = [V] # DFS 
visited = [0 for _ in range(N+1)] #  .
while stack:
    current=stack.pop() #  
    if visited[current]==1: #   pass
        continue
    sequence_dfs.append(current)
    visited[current] = 1
    for n in reversed(sorted(graph[current])):
        if visited[n]==1: #   
            continue
        stack.append(n)
print(*sequence_dfs)
print(*sequence_bfs)
```
