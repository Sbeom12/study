```python
from collections import deque
N, M = map(int, input().split())
graph = [list(map(int, str(input()))) for _ in range(N)]

queue = deque()
queue.append([0,0,0])
visited = [[-1 for _ in range(M)] for _ in range(N)]
visited[0][0] =1
move_sequence = [[1,0], [-1, 0], [0,1], [0, -1]]
first_time = 1
while queue:
    x, y, wall = queue.popleft()
    if visited[-1][-1] != -1 or visited[0][0]> N+M:
        break
    for sequence in move_sequence:
        nx , ny = x +sequence[0], y+sequence[1]
        if nx < 0 or nx>= M or ny>= N or ny < 0:
            continue
        if visited[ny][nx] == -1:
            if graph[ny][nx] == 0:
                visited[ny][nx] = visited[y][x]+1
                queue.append([nx, ny, wall])
            else: # 벽 부시기
                if wall == 0:
                    visited[ny][nx] = visited[y][x]+1
                    queue.append([nx, ny, 1])
        else:
            if wall == 0:
                if graph[ny][nx] == 0:
                    visited[ny][nx] = visited[y][x]+1
                    queue.append([nx, ny, wall])
                else:
                    visited[ny][nx] = visited[y][x]+1
                    queue.append([nx, ny, wall+1])

print(visited[-1][-1])
```
