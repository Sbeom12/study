# 단지 번호 붙이기
---


## 풀이
* 시간 제한 1초 : 계산량 대충 2000만 정도
* 메모리 제한 128MB
* 방법: 완전 탐색 BFS or DFS
    * DFS
        * 구현 방법 : stack or 재귀함수
    * BFS
        * 구현 방법 : queue

```python
n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input())))
count_list =[]

def dfs(x,y, value):
    if x<= -1 or x>= n or y <= -1 or y >= n:
        return False
    if graph[x][y] == 1:
        graph[x][y] = value
        dfs(x-1, y, value)
        dfs(x, y+1, value)
        dfs(x+1, y, value)
        dfs(x, y-1, value)
        return True
    return False
result = 2
for i in range(n):
    for j in range(n):
        if dfs(i, j, result) == True:
            result += 1
print(result-2)
for i in range(2, result):
    count = 0
    for line in graph:
        temp = line.count(i)
        count += temp
    count_list.append(count)
for num in sorted(count_list):
    print(num)
```

## 검산: 
---
* N = 100,000
* M = 200,000
* 정렬사용시 : $O(Nlog(N))$  
* 인접행렬시 : $O(n^2)$  
* 인접리스트 : $O(N+M)$  

* 인접리스트 시 통과 가능.
