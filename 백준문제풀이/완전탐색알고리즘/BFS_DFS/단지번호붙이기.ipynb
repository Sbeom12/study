*   1 :   2000 
*   128MB
* :   BFS or DFS
    * DFS
        *   : stack or 
    * BFS
        *   : queue

```python
n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input())))
count_list =[]

def dfs(x,y, value):
    if x<= -1 or x>= n or y <= -1 or y >= n:
        return False
    if graph[x][y] == 1:
        graph[x][y] = value
        dfs(x-1, y, value)
        dfs(x, y+1, value)
        dfs(x+1, y, value)
        dfs(x, y-1, value)
        return True
    return False
result = 2
for i in range(n):
    for j in range(n):
        if dfs(i, j, result) == True:
            result += 1
print(result-2)
for i in range(2, result):
    count = 0
    for line in graph:
        temp = line.count(i)
        count += temp
    count_list.append(count)
for num in sorted(count_list):
    print(num)
```
