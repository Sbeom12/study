```python
from collections import deque
T = int(input())
for _ in range(T):
    i = int(input())
    start_x, start_y = map(int, input().split())
    target_x, target_y = map(int, input().split())
    move_case = [(-2, 1), (-2, -1), (2, 1), (2, -1), (1, 2),(-1, 2), (1, -2), (-1, -2)]
    queue = deque() # 초기 위치 넣기
    graph =[[0 for _ in range(i)] for _ in range(i)]
    queue.append((start_x, start_y))
    while queue:
        x, y = queue.popleft()
        if x == target_x and y == target_y:
            print(graph[y][x])
            break 
        for move in move_case:
            nx, ny = x+move[0], y+move[1]
            if nx < 0 or ny < 0 or nx>= i or ny>=i: # 체스판을 벗어나는 경우 패스~
                continue
            if graph[ny][nx]== 0:
                graph[ny][nx] = graph[x][y] + 1
                queue.append((nx, ny))
        #print(graph)
            
```

* 이진 탐색이 아니면 시간초과 발생(input 사용시)
* N = 500,000
* M = 500,000

```python
K, N = map(int, input().split())
cable_list = [int(input()) for _ in range(K)]
cable_list.sort()

def binary_search(target, start, end):
    answer = 0
    while start<=end:
        mid = (start+end)//2
        if mid == 0:
            mid =1
        temp = 0
        for cable in cable_list:
            temp+= cable//mid
        if temp >= target:
            start = mid +1
            answer = mid
        else:
            end = mid - 1
    return answer
print(binary_search(N, 0, max(cable_list)))

```

```python
import sys
N, M = map(int, input().split())
tree_list = list(map(int, input().split()))

# 나무의 높이를 오름차순으로 정렬
tree_list.sort()

def binary_search(target, start, end):
    answer = 0
    while start <= end:
        mid = (start + end) // 2
        temp = sum([tree - mid for tree in tree_list if tree > mid])

        if temp >= target:
            if mid > answer:
                answer = mid
            start = mid + 1
        else:
            end = mid - 1

    return answer

print(binary_search(M, 0, max(tree_list)))
```
