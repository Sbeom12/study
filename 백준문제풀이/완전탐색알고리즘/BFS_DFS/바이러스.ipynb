## 풀이
* 시간 제한 1초 : 계산량 대충 2000만 정도
* 메모리 제한 128MB
* DFS
    * 구현 방법 : stack or 재귀함수
* BFS
    * 구현 방법 : queue

```python
from collections import defaultdict, deque
N = int(input())
E = int(input())
graph = defaultdict(set)
for _ in range(E):
    node1, node2 = map(int, input().split())
    graph[node1].add(node2)
    graph[node2].add(node1) 
visited = [0 for _ in range(N+1)] # 방문 확인.
queue = deque()
queue.append(1) # 무조건 1부터 시작.
while queue:
    current=queue.popleft() # 현재 방문
    if visited[current] == 1: # 방문 했다면 pass
        continue
    visited[current] = 1
    for n in sorted(graph[current]):
        if visited[n] == 1: # 이전에 방문햇으면 통과
            continue
        queue.append(n)
answer = sum(visited)-1
print(answer)
```
