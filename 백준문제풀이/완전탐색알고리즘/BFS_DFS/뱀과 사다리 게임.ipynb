```python
from collections import deque
N, M = map(int, input().split())
snake = {}
for _ in range(N+M):
    x1, x2 = map(int, input().split())
    snake[x1] = x2
        
graph = [0 for _ in range(101)]
queue = deque()

queue.append(1)
while queue:
    x = queue.popleft()
    for i in range(1,7):
        nx = x + i
        if nx>= 101: #    ~
            continue
        if nx in snake:
            nx = snake[nx]
        if graph[nx] == 0:
            graph[nx] = graph[x] + 1
            queue.append(nx)
print(graph[-1])
```
