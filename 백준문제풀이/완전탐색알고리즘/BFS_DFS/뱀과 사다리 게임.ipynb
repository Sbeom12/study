## 풀이
* 시간 제한 : 1초
* 메모리 제한 : 512MB
* 풀이 : 최소 거리 = BFS
    * 큐로 구현.

```python
from collections import deque
N, M = map(int, input().split())
snake = {}
for _ in range(N+M):
    x1, x2 = map(int, input().split())
    snake[x1] = x2
        
graph = [0 for _ in range(101)]
queue = deque()

queue.append(1)
while queue:
    x = queue.popleft()
    for i in range(1,7):
        nx = x + i
        if nx>= 101: # 체스판을 벗어나는 경우 패스~
            continue
        if nx in snake:
            nx = snake[nx]
        if graph[nx] == 0:
            graph[nx] = graph[x] + 1
            queue.append(nx)
print(graph[-1])
```
