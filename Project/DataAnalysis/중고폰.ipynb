```python
data_list =[]
for i in range(1, 11):
    page_num = i
    url = 'https://snphone.co.kr/shop/?SM=list&company_code=&buy_search=&buy_price=&buy_color=&buy_hashtag=&buy_type=&buy_order=&buy_select_key=&buy_keyword=&p=' + str(page_num)
    brower.get(url)
    soup = BeautifulSoup(brower.page_source, 'html.parser')
    data_list += soup.select('.prd_area > .prd_info')
    time.sleep(5)
```

```python
name_list, price_list, prdtag_list, color_list =[],[],[],[]
for data in data_list:
    name = data.select('h3.name')[0].get_text()
    price= data.select('.price_line > b')[0].get_text()
    len_color = len(data.select('.color_kind > span'))
    prd_link = 'https://snphone.co.kr/' + data.find('a')['href']
    name_list.append(name)
    price_list.append(price)
    color_list.append(len_color)
    prdtag_list.append(prd_link)
    print(prd_link)
```
